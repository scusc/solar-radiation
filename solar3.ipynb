```

import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```
dataset = pd.read_csv('solatdata.csv')
dataset = dataset.sort_values(['UNIXTime'], ascending = [True])
test_data=pd.read_csv('TEST.csv')
dataset.head()
```

```
from datetime import datetime
from pytz import timezone
import pytz
hawaii= timezone('Pacific/Honolulu')
dataset.index =  pd.to_datetime(dataset['UNIXTime'], unit='s')
dataset.index = dataset.index.tz_localize(pytz.utc).tz_convert(hawaii)
dataset['MonthOfYear'] = dataset.index.strftime('%m').astype(int)
dataset['DayOfYear'] = dataset.index.strftime('%j').astype(int)
dataset['WeekOfYear'] = dataset.index.strftime('%U').astype(int)
dataset['TimeOfDay(h)'] = dataset.index.hour
dataset['TimeOfDay(m)'] = dataset.index.hour*60 + dataset.index.minute
dataset['TimeOfDay(s)'] = dataset.index.hour*60*60 + dataset.index.minute*60 + dataset.index.second
dataset['TimeSunRise'] = pd.to_datetime(dataset['TimeSunRise'], format='%H:%M:%S')
dataset['TimeSunSet'] = pd.to_datetime(dataset['TimeSunSet'], format='%H:%M:%S')
dataset['DayLength(s)'] = dataset['TimeSunSet'].dt.hour*60*60 \
                           + dataset['TimeSunSet'].dt.minute*60 \
                           + dataset['TimeSunSet'].dt.second \
                           - dataset['TimeSunRise'].dt.hour*60*60 \
                           - dataset['TimeSunRise'].dt.minute*60 \
                           - dataset['TimeSunRise'].dt.second
dataset.head()
```

```
test_data=pd.read_csv('TEST.csv')
hawaii= timezone('Pacific/Honolulu')
test_data.index =  pd.to_datetime(test_data['UNIXTime'], unit='s')
test_data.index = test_data.index.tz_localize(pytz.utc).tz_convert(hawaii)
test_data['MonthOfYear'] = test_data.index.strftime('%m').astype(int)
test_data['DayOfYear'] = test_data.index.strftime('%j').astype(int)
test_data['WeekOfYear'] = test_data.index.strftime('%U').astype(int)
test_data['TimeOfDay(h)'] = test_data.index.hour
test_data['TimeOfDay(m)'] = test_data.index.hour*60 + test_data.index.minute
test_data['TimeOfDay(s)'] = test_data.index.hour*60*60 + test_data.index.minute*60 + test_data.index.second
test_data['TimeSunRise'] = pd.to_datetime(test_data['TimeSunRise'], format='%H:%M:%S')
test_data['TimeSunSet'] = pd.to_datetime(test_data['TimeSunSet'], format='%H:%M:%S')
test_data['DayLength(s)'] = test_data['TimeSunSet'].dt.hour*60*60 \
                           + test_data['TimeSunSet'].dt.minute*60 \
                           + test_data['TimeSunSet'].dt.second \
                           - test_data['TimeSunRise'].dt.hour*60*60 \
                           - test_data['TimeSunRise'].dt.minute*60 \
                           - test_data['TimeSunRise'].dt.second
```

```

dataset['HoursSunWasAvailable']=pd.to_datetime(dataset['TimeSunSet'])-pd.to_datetime(dataset['TimeSunRise'])
test_data['HoursSunWasAvailable']=pd.to_datetime(test_data['TimeSunSet'])-pd.to_datetime(test_data['TimeSunRise'])


dataset['HoursSunWasAvailable']=dataset['HoursSunWasAvailable'].apply(lambda x:str(x).split(' ')[2])
test_data['HoursSunWasAvailable']=test_data['HoursSunWasAvailable'].apply(lambda x:str(x).split(' ')[2])


dataset['HoursSunWasAvailable']=dataset['HoursSunWasAvailable'].apply(lambda x:x.split(':')[0])
test_data['HoursSunWasAvailable']=test_data['HoursSunWasAvailable'].apply(lambda x:x.split(':')[0])


dataset['HoursSunWasAvailable']=dataset['HoursSunWasAvailable'].astype(int)
test_data['HoursSunWasAvailable']=test_data['HoursSunWasAvailable'].astype(int)


dataset.drop(['Data','TimeSunSet','TimeSunRise'],axis=1,inplace=True)
test_data.drop(['Data','TimeSunSet','TimeSunRise'],axis=1,inplace=True)
```

```
dataset.drop(['UNIXTime','Time'],axis=1,inplace=True)
test_data.drop(['idx','UNIXTime','Time'],axis=1,inplace=True)
```
